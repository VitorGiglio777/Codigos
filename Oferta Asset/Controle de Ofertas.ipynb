```python

# Baixa a base CVM

import os
import zipfile
import requests
from datetime import datetime

# Caminho da pasta
caminho_pasta = r"C:\\SP\\REAG DISTRIBUIDORA DE TITULOS E VALORES MOBILIARIOS S.A\\OPERAÇÕES - Documentos\\Ofertas e Cadastros\\1 - Asset\\8 - Bases Python\\0 - Base Geral\\00 - Bases CVM\\1 - Ofertas ICVM 160\\"

# Função para excluir arquivos antigos
def excluir_arquivos_antigos(caminho_pasta):
    if os.path.exists(caminho_pasta):
        arquivos = os.listdir(caminho_pasta)
        for arquivo in arquivos:
            caminho_arquivo = os.path.join(caminho_pasta, arquivo)
            try:
                if os.path.isfile(caminho_arquivo):
                    os.remove(caminho_arquivo)
                    print(f"Arquivo excluído: {arquivo}")
            except Exception as e:
                print(f"Erro ao excluir {arquivo}: {e}")
    else:
        print("A pasta especificada não existe.")

# Função para baixar o arquivo ZIP
def download_file():
    url = "https://dados.cvm.gov.br/dados/OFERTA/DISTRIB/DADOS/oferta_distribuicao.zip"
    file_name = f"registro_fundo_classe_{datetime.now().strftime('%Y%m%d_%H%M%S')}.zip"
    file_path = os.path.join(caminho_pasta, file_name)

    try:
        response = requests.get(url)
        response.raise_for_status()
        with open(file_path, 'wb') as file:
            file.write(response.content)
        print(f"Arquivo baixado com sucesso: {file_path}")
    except Exception as e:
        print(f"Erro ao baixar o arquivo: {e}")

# Função para extrair arquivos CSV do ZIP
def extrair_primeiro_zip(caminho_pasta):
    for arquivo in os.listdir(caminho_pasta):
        if arquivo.endswith('.zip'):
            caminho_arquivo_zip = os.path.join(caminho_pasta, arquivo)
            print(f"Arquivo ZIP encontrado: {arquivo}")
            with zipfile.ZipFile(caminho_arquivo_zip, 'r') as zip_ref:
                for item in zip_ref.namelist():
                    if item.endswith('.csv'):
                        destino_arquivo = os.path.join(caminho_pasta, os.path.basename(item))
                        with open(destino_arquivo, 'wb') as f:
                            f.write(zip_ref.read(item))
                        print(f"Arquivo extraído: {destino_arquivo}")
            return
    print("Nenhum arquivo ZIP encontrado na pasta.")

# Executar as funções manualmente
excluir_arquivos_antigos(caminho_pasta)
download_file()
extrair_primeiro_zip(caminho_pasta)
```

```python
#LÓGICA DE RESULTADOS DA COLUNA STATUS ATUAL 

import pandas as pd
from datetime import datetime

# Criando uma cópia do DataFrame original
df_logica = df_numero_requerimento.copy()

# Obtendo a data atual no formato date
data_atual = datetime.now().date()

# Função para determinar o STATUS_ATUAL
def determinar_status(row):
    status_requerimento = row['Status_Requerimento']
    alerta_encerramento = row['ALERTA_ENCERRAMENTO']
    prazo_encerramento_final = row['PRAZO_ENCERRAMENTO_FINAL']
    data_encerramento = row['Data_Encerramento']
    
    if status_requerimento == 'Registro Concedido':
        if data_atual < alerta_encerramento:
            return 'OFERTA OK'
        elif alerta_encerramento <= data_atual <= prazo_encerramento_final:
            return 'ALERTA DE ENCERRAMENTO'
        elif data_atual > prazo_encerramento_final and pd.isna(data_encerramento):
            return 'OFERTA VENCIDA'
    elif status_requerimento == 'Registro Caducado':
        if pd.isna(data_encerramento):
            return 'OFERTA CADUCADA SEM DATA DE ENCERRAMENTO'
        elif data_encerramento > datetime(1990, 1, 1).date():
            return 'OFERTA ENCERRADA'
    elif status_requerimento == 'Oferta Encerrada':
        return 'OFERTA ENCERRADA'
    return None

# Convertendo as colunas para datetime.date, se necessário
df_logica['ALERTA_ENCERRAMENTO'] = pd.to_datetime(df_logica['ALERTA_ENCERRAMENTO']).dt.date
df_logica['PRAZO_ENCERRAMENTO_FINAL'] = pd.to_datetime(df_logica['PRAZO_ENCERRAMENTO_FINAL']).dt.date
df_logica['Data_Encerramento'] = pd.to_datetime(df_logica['Data_Encerramento'], errors='coerce').dt.date

# Aplicando a lógica para determinar o STATUS_ATUAL
df_logica['STATUS_ATUAL'] = df_logica.apply(determinar_status, axis=1)

# Exibindo o DataFrame resultante
display(df_logica)


```

```python
import pandas as pd
import re
import os
from IPython.display import display

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

def setup_driver():
    """
    Configura o ChromeDriver apenas uma vez, antes de processar todas as URLs.
    """
    options = Options()
    options.add_argument("--headless")       # Executar o navegador em modo headless
    options.add_argument("--disable-gpu")
    
    driver = webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=options
    )
    
    # Define tempo máximo de 1 segundo para encontrar elementos antes de lançar exceção
    driver.implicitly_wait(1)
    
    return driver

def get_xpaths(driver, url, xpath_queries):
    """
    Usa o driver já aberto para acessar a URL e extrair o texto de cada XPATH.
    """
    driver.get(url)

    data = []
    # Aqui está usando range(1, 5) como exemplo para 4 linhas;
    # ajuste se precisar de mais (ou menos) linhas.
    for i in range(1, 5):
        row_data = []
        for xpath_query in xpath_queries:
            try:
                modified_xpath = xpath_query.replace("[1]", f"[{i}]")
                element = driver.find_element(By.XPATH, modified_xpath)
                row_data.append(element.text)
            except Exception:
                # Se não encontrar o elemento (ou ocorrer outro erro), adiciona string vazia
                row_data.append("")
        
        data.append(row_data)

    column_names = [f"Coluna {i + 1}" for i in range(len(xpath_queries))]
    df = pd.DataFrame(data, columns=column_names)
    return df

def extract_numbers(value):
    """
    Extrai apenas dígitos de uma string, retornando uma string com os números concatenados.
    Se não houver números, retorna string vazia.
    """
    numbers = re.findall(r'\d+', value)
    return "".join(numbers) if numbers else ""

def comunicado_mercado_logic(row):
    """
    Lógica para preencher a coluna COMUNICADO_MERCADO.
    """
    if "Comunicado ao Mercado" in row["Coluna 4"] and row["Números da Coluna 1"] != "00":
        return "Comunicado ao Mercado Ok"
    else:
        return "Falta Comunicado ao Mercado"

def comunicado_inicio_distribuicao_logic(row):
    """
    Lógica para preencher a coluna COMUNICADO_INICIO_DISTRIBUIÇÃO.
    """
    if "Anúncio de Início da Distribuição" in row["Coluna 4"] and row["Números da Coluna 1"] != "00":
        return "Anúncio de Início da Distribuição Ok"
    else:
        return "Falta Anúncio de Início da Distribuição"

def comunicado_mercado_final_logic(df):
    """
    Verifica se há pelo menos uma linha com 'Comunicado ao Mercado Ok'.
    Caso exista, insere 'Comunicado ao Mercado Verificado' na primeira linha;
    caso contrário, 'Comunicado ao Mercado Incompleto'.
    """
    if (df["COMUNICADO_MERCADO"] == "Comunicado ao Mercado Ok").any():
        return ["COMUNICADO AO MERCADO VERIFICADO"] + [""] * (len(df) - 1)
    else:
        return ["COMUNICADO AO MERCADO INCOMPLETO"] + [""] * (len(df) - 1)

def comunicado_inicio_distribuicao_final_logic(df):
    """
    Verifica se há pelo menos uma linha com 'Anúncio de Início da Distribuição Ok'.
    Caso exista, insere 'Anúncio de Início da Distribuição Verificado' na primeira linha;
    caso contrário, 'Anúncio de Início da Distribuição Incompleto'.
    """
    if (df["COMUNICADO_INICIO_DISTRIBUIÇÃO"] == "Anúncio de Início da Distribuição Ok").any():
        return ["ANUNCIO DE INÍCIO DA DISTRIBUIÇÃO VERIFICADO"] + [""] * (len(df) - 1)
    else:
        return ["ANUNCIO DE INÍCIO DA DISTRIBUIÇÃO INCOMPLETO"] + [""] * (len(df) - 1)

def numero_para_romano(num):
    """
    Converte um número inteiro de 1 a 10 para algarismo romano.
    Caso esteja fora do range, retorna o número como string normal.
    """
    romanos = {
        1: "I", 2: "II", 3: "III", 4: "IV", 5: "V",
        6: "VI", 7: "VII", 8: "VIII", 9: "IX", 10: "X"
    }
    return romanos.get(num, str(num))


# =============================================================================
#                 PROGRAMA PRINCIPAL - Exemplo de uso
# =============================================================================

# Exemplo fictício de df_url_ofertas
# Ajuste conforme seus dados:
# df_url_ofertas = pd.DataFrame({
#     "URL das ofertas": [
#         "https://web.cvm.gov.br/sre-publico-cvm/#/oferta-publica/18068",
#         "https://web.cvm.gov.br/sre-publico-cvm/#/oferta-publica/99999"
#     ]
# })

# Defina aqui seus XPaths
xpath_queries = [
    '//*[@id="container"]/div/div/div/sre-documentos/div/div/fieldset/div/div/table/tbody/tr[1]/td[6]',
    '//*[@id="container"]/div/div/div/sre-documentos/div/div/fieldset/div/div/table/tbody/tr[1]/td[5]',
    '//*[@id="container"]/div/div/div/sre-documentos/div/div/fieldset/div/div/table/tbody/tr[1]/td[4]',
    '//*[@id="container"]/div/div/div/sre-documentos/div/div/fieldset/div/div/table/tbody/tr[1]/td[3]',
    '//*[@id="container"]/div/div/div/sre-participantes/div/div[1]/div/fieldset/div[2]/div',
    '//*[@id="container"]/div/div/div/sre-informacoes-gerais/div/div/fieldset/div[6]/div/label'
]

# 1) Inicializa o driver uma única vez
driver = setup_driver()

# 2) Itera sobre as URLs do DataFrame
for index, row in df_url_ofertas.iterrows():
    url = row["URL das ofertas"]
    print(f"Processando index {index} - URL: {url}")

    # Chama a função para extrair dados de cada página
    result_df = get_xpaths(driver, url, xpath_queries)

    # Renomeia colunas 5 e 6
    result_df.rename(columns={"Coluna 5": "CNPJ_DOC"}, inplace=True)
    result_df.rename(columns={"Coluna 6": "PROCESSO_DOC"}, inplace=True)

    # Limpa a coluna "CNPJ_DOC"
    result_df["CNPJ_DOC"] = (result_df["CNPJ_DOC"]
                             .str.replace("CNPJ/CPF:", "", regex=False)
                             .str.replace("CNPJ:", "", regex=False)
                             .str.strip())

    # Limpa a coluna "PROCESSO_DOC"
    result_df["PROCESSO_DOC"] = (result_df["PROCESSO_DOC"]
                                 .str.replace("Nº do Processo:", "", regex=False)
                                 .str.strip())

    # Cria coluna com apenas números extraídos da "Coluna 1"
    result_df["Números da Coluna 1"] = result_df["Coluna 1"].apply(extract_numbers)

    # Aplica lógicas de comunicado
    result_df["COMUNICADO_MERCADO"] = result_df.apply(comunicado_mercado_logic, axis=1)
    result_df["COMUNICADO_INICIO_DISTRIBUIÇÃO"] = result_df.apply(comunicado_inicio_distribuicao_logic, axis=1)

    result_df["COMUNICADO_MERCADO_FINAL"] = comunicado_mercado_final_logic(result_df)
    result_df["COMUNICADO_INICIO_DISTRIBUIÇÃO_FINAL"] = comunicado_inicio_distribuicao_final_logic(result_df)

    # Exibe o DataFrame (opcional)
    display(result_df)

    # Gera um nome-base para o arquivo Excel, baseado na primeira linha de "CNPJ_DOC"
    try:
        nome_base = result_df["CNPJ_DOC"].iloc[0].strip()
    except IndexError:
        nome_base = "dados_cvm"

    # Remove caracteres inválidos do nome do arquivo
    nome_base = re.sub(r'[\\/:"*?<>|]+', '', nome_base)

    # Define o nome padrão do Excel
    nome_arquivo_excel = f"{nome_base}.xlsx"
    contador_sufixo = 1

    # Define a pasta de destino
    pasta_destino = r"C:\\SP\\REAG DISTRIBUIDORA DE TITULOS E VALORES MOBILIARIOS S.A\\OPERAÇÕES - Documentos\\Ofertas e Cadastros\\1 - Asset\\8 - Bases Python\\1 - Controle de Ofertas\\BASE DOCUMENTOS\\"  # <-- Ajuste aqui!

    # Cria o caminho completo para o arquivo Excel
    caminho_completo_arquivo = os.path.join(pasta_destino, nome_arquivo_excel)

    # Caso já exista, vamos incrementando sufixos romanos
    while os.path.exists(caminho_completo_arquivo):
        sufixo_romano = numero_para_romano(contador_sufixo)
        nome_arquivo_excel = f"{nome_base} {sufixo_romano}.xlsx"
        caminho_completo_arquivo = os.path.join(pasta_destino, nome_arquivo_excel)
        contador_sufixo += 1

    # Salva o DataFrame em Excel
    with pd.ExcelWriter(caminho_completo_arquivo, engine='openpyxl') as writer:
        result_df.to_excel(writer, sheet_name='Dados', index=False)

    print(f"Excel gerado: {caminho_completo_arquivo}")

# 3) Finaliza o driver após processar TODAS as URLs
driver.quit()

print("\n\nProcessamento concluído para todas as URLs.")

```
