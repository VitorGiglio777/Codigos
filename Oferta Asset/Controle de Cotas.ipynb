```python

# Baixa a base CVM COM TODAS AS OFERTAS 

import os
import zipfile
import requests
from datetime import datetime

# Caminho da pasta
caminho_pasta = r"C:\\SP\\REAG DISTRIBUIDORA DE TITULOS E VALORES MOBILIARIOS S.A\\OPERAÇÕES - Documentos\\Ofertas e Cadastros\\1 - Asset\\8 - Bases Python\\0 - Base Geral\\00 - Bases CVM\\1 - Ofertas ICVM 160\\"

# Função para excluir arquivos antigos
def excluir_arquivos_antigos(caminho_pasta):
    if os.path.exists(caminho_pasta):
        arquivos = os.listdir(caminho_pasta)
        for arquivo in arquivos:
            caminho_arquivo = os.path.join(caminho_pasta, arquivo)
            try:
                if os.path.isfile(caminho_arquivo):
                    os.remove(caminho_arquivo)
                    print(f"Arquivo excluído: {arquivo}")
            except Exception as e:
                print(f"Erro ao excluir {arquivo}: {e}")
    else:
        print("A pasta especificada não existe.")

# Função para baixar o arquivo ZIP
def download_file():
    url = "https://dados.cvm.gov.br/dados/OFERTA/DISTRIB/DADOS/oferta_distribuicao.zip"
    file_name = f"registro_fundo_classe_{datetime.now().strftime('%Y%m%d_%H%M%S')}.zip"
    file_path = os.path.join(caminho_pasta, file_name)

    try:
        response = requests.get(url)
        response.raise_for_status()
        with open(file_path, 'wb') as file:
            file.write(response.content)
        print(f"Arquivo baixado com sucesso: {file_path}")
    except Exception as e:
        print(f"Erro ao baixar o arquivo: {e}")

# Função para extrair arquivos CSV do ZIP
def extrair_primeiro_zip(caminho_pasta):
    for arquivo in os.listdir(caminho_pasta):
        if arquivo.endswith('.zip'):
            caminho_arquivo_zip = os.path.join(caminho_pasta, arquivo)
            print(f"Arquivo ZIP encontrado: {arquivo}")
            with zipfile.ZipFile(caminho_arquivo_zip, 'r') as zip_ref:
                for item in zip_ref.namelist():
                    if item.endswith('.csv'):
                        destino_arquivo = os.path.join(caminho_pasta, os.path.basename(item))
                        with open(destino_arquivo, 'wb') as f:
                            f.write(zip_ref.read(item))
                        print(f"Arquivo extraído: {destino_arquivo}")
            return
    print("Nenhum arquivo ZIP encontrado na pasta.")

# Executar as funções manualmente
excluir_arquivos_antigos(caminho_pasta)
download_file()
extrair_primeiro_zip(caminho_pasta)
```

```python
import pandas as pd

# Cria uma cópia do DataFrame df_final
df_quanticotas = df_final.copy()

# Define a lógica para a coluna 'QUANTIDADE DISPONIVEL'
def calcular_quantidade_disponivel(row):
    if row['TIPO_CONTROLE'] == 'FINANCEIRO':
        return 'CONTROLE POR FINANCEIRO'
    elif row['SUBCLASSE'] == 'Junior':
        return row['QUANT_JNR_INTEGRALIZADA_ANTES_EMISSÃO_ATUAL'] - row['QUANTIDADE DE COTA'] + row['QUANTI_TOTAL']
    elif row['ID META_BASE'] in [0, 1]:
        return row['QUANTI_TOTAL']
    else:
        return row['QUANTI_EMITIDAS'] - row['QUANTIDADE DE COTA']

# Aplica a função ao DataFrame
df_quanticotas['QUANTIDADE DISPONIVEL'] = df_quanticotas.apply(calcular_quantidade_disponivel, axis=1)

# Exibe o DataFrame resultante
display(df_quanticotas)


```

```python
import pandas as pd

# Cria uma cópia do DataFrame
df_porcentagem = df_quanticotas.copy()

# Converte apenas os valores que não são "CONTROLE POR FINANCEIRO"
df_porcentagem['QUANTIDADE DISPONIVEL'] = df_porcentagem['QUANTIDADE DISPONIVEL'].apply(
    lambda x: pd.to_numeric(x, errors='coerce') if x != 'CONTROLE POR FINANCEIRO' else x
)

# Converte QUANTI_TOTAL normalmente
df_porcentagem['QUANTI_TOTAL'] = pd.to_numeric(df_porcentagem['QUANTI_TOTAL'], errors='coerce')

# Ajusta a lógica para criar a coluna '% DISPONIVEL'
def calcular_percentual(row):
    if row['TIPO_CONTROLE'] == 'FINANCEIRO':
        return 'CONTROLE POR FINANCEIRO'
    elif isinstance(row['QUANTIDADE DISPONIVEL'], (int, float)) and pd.notnull(row['QUANTI_TOTAL']) and row['QUANTI_TOTAL'] != 0:
        return f"{round((row['QUANTIDADE DISPONIVEL'] / row['QUANTI_TOTAL']) * 100, 2)}%"
    else:
        return None  # Ou 'ND', 'N/A' etc., conforme necessidade

# Aplica a função ao DataFrame
df_porcentagem['% DISPONIVEL'] = df_porcentagem.apply(calcular_percentual, axis=1)

# Exibe o DataFrame resultante
display(df_porcentagem)


```

```python
import pandas as pd

# Cria uma cópia do DataFrame original
df_financeiro_disponivel = df_porcentagem.copy()

# Converte VALOR COTA para numérico
df_financeiro_disponivel['VALOR COTA'] = pd.to_numeric(df_financeiro_disponivel['VALOR COTA'], errors='coerce')

# Função para calcular FINANCEIRO DISPONIVEL
def calcular_financeiro(row):
    try:
        if row['ID META_BASE'] == 1:
            return row['FINANCEIRO']
        elif str(row['QUANTIDADE DISPONIVEL']).strip().upper() == 'CONTROLE POR FINANCEIRO':
            return 'CONTROLE POR FINANCEIRO'
        else:
            qtd = pd.to_numeric(row['QUANTIDADE DISPONIVEL'], errors='coerce')
            valor = row['VALOR COTA']
            return qtd * valor
    except:
        return None

# Aplica a função para calcular o financeiro
df_financeiro_disponivel['FINANCEIRO DISPONIVEL'] = df_financeiro_disponivel.apply(calcular_financeiro, axis=1)

# Garante que a coluna 'DATA CARTEIRA' existe
if 'DATA CARTEIRA' not in df_financeiro_disponivel.columns:
    df_financeiro_disponivel['DATA CARTEIRA'] = None

# Atualiza a coluna 'DATA CARTEIRA' com "PRIMEIRO APORTE" onde ID META_BASE == 1
df_financeiro_disponivel.loc[df_financeiro_disponivel['ID META_BASE'] == 1, 'DATA CARTEIRA'] = 'PRIMEIRO APORTE'

# Converte a coluna de data
df_financeiro_disponivel['Data_Registro'] = pd.to_datetime(df_financeiro_disponivel['Data_Registro'], errors='coerce')

# Calcula ENCERRAMENTO COTA com exceção para "PRIVADA - DISTRIBUIÇÃO"
df_financeiro_disponivel['ENCERRAMENTO COTA'] = df_financeiro_disponivel.apply(
    lambda row: 'OFERTA PRIVADA' if str(row['NUMERO_SRE']).strip().upper() == 'PRIVADA - DISTRIBUIÇÃO'
    else (row['Data_Registro'] + pd.Timedelta(days=180)).strftime('%d/%m/%Y') if pd.notnull(row['Data_Registro'])
    else None,
    axis=1
)

# Remove colunas duplicadas, se existirem
df_financeiro_disponivel = df_financeiro_disponivel.loc[:, ~df_financeiro_disponivel.columns.duplicated()]

# Exibe o resultado final
display(df_financeiro_disponivel)

```

```python
# Cria cópias independentes dos DataFrames originais
OFERTAS_ATIVAS_COPIA = OFERTAS_ATIVAS.copy()
OFERTAS_ENCERRADAS_COPIA = OFERTAS_ENCERRADAS.copy()

# Lista de colunas a formatar com separador de milhar
colunas_formatar = [
    'QUANTIDADE DISPONIVEL',
    'QUANTI_TOTAL',
    'FINANCEIRO',
    'FINANCEIRO DISPONIVEL'
]

# Função de formatação com separador de milhar brasileiro e duas casas decimais
def formatar_milhar_brasileiro(valor):
    if pd.notnull(valor) and isinstance(valor, (int, float)):
        return '{:,.2f}'.format(valor).replace(',', 'X').replace('.', ',').replace('X', '.')
    return valor

# Aplica a formatação nas cópias
for coluna in colunas_formatar:
    if coluna in OFERTAS_ATIVAS_COPIA.columns:
        OFERTAS_ATIVAS_COPIA[coluna] = OFERTAS_ATIVAS_COPIA[coluna].apply(formatar_milhar_brasileiro)
    if coluna in OFERTAS_ENCERRADAS_COPIA.columns:
        OFERTAS_ENCERRADAS_COPIA[coluna] = OFERTAS_ENCERRADAS_COPIA[coluna].apply(formatar_milhar_brasileiro)

# Formata a coluna DATA CARTEIRA para DD/MM/AAAA
def formatar_data(data):
    try:
        return pd.to_datetime(data).strftime('%d/%m/%Y')
    except:
        return data  # retorna como está se não for data válida

if 'DATA CARTEIRA' in OFERTAS_ATIVAS_COPIA.columns:
    OFERTAS_ATIVAS_COPIA['DATA CARTEIRA'] = OFERTAS_ATIVAS_COPIA['DATA CARTEIRA'].apply(formatar_data)

if 'DATA CARTEIRA' in OFERTAS_ENCERRADAS_COPIA.columns:
    OFERTAS_ENCERRADAS_COPIA['DATA CARTEIRA'] = OFERTAS_ENCERRADAS_COPIA['DATA CARTEIRA'].apply(formatar_data)

# Exibe os DataFrames copiados e formatados
display(OFERTAS_ATIVAS_COPIA)
display(OFERTAS_ENCERRADAS_COPIA)



```

```python
import os
import pandas as pd

# Lista de caminhos onde o arquivo será salvo
caminhos_arquivos = [
    r"C:\SP\REAG DISTRIBUIDORA DE TITULOS E VALORES MOBILIARIOS S.A\OPERAÇÕES - Documentos\Ofertas e Cadastros\1 - Asset\5 - Controles\3 - Controles de Cotas\Controle de Cotas.xlsx",
    r"C:\SP\REAG DISTRIBUIDORA DE TITULOS E VALORES MOBILIARIOS S.A\SALES - SALES - Documentos\Documentos\Controles\Controle de Cotas\Controle de Cotas.xlsx",
]

# DataFrames e nomes das abas
abas = [
    (OFERTAS_ATIVAS_COPIA, "OFERTAS ATIVAS"),
    (OFERTAS_ENCERRADAS_COPIA, "OFERTAS ENCERRADAS"),
]

# Colunas a alinhar à direita
colunas_direita = {
    "COTA","SERIE","STATUS","QUANTI_TOTAL","FINANCEIRO",
    "QUANTIDADE DISPONIVEL","FINANCEIRO DISPONIVEL","% DISPONIVEL",
    "ENCERRAMENTO COTA","DATA CARTEIRA","OBS",
}

def salvar_excel(caminho_arquivo):
    # Garante que a pasta existe
    os.makedirs(os.path.dirname(caminho_arquivo), exist_ok=True)

    with pd.ExcelWriter(caminho_arquivo, engine="xlsxwriter") as writer:
        for df, nome_aba in abas:
            df = pd.DataFrame(df)
            df.to_excel(writer, sheet_name=nome_aba, index=False)

            workbook  = writer.book
            worksheet = writer.sheets[nome_aba]

            # formatos
            header_fmt = workbook.add_format({
                "bold": True, "font_color": "white", "bg_color": "#054605",
                "align": "center", "valign": "vcenter",
            })
            right_fmt = workbook.add_format({"align": "right"})

            # cabeçalho pintado apenas onde há texto
            for col_idx, col_name in enumerate(df.columns):
                if str(col_name).strip():
                    worksheet.write(0, col_idx, col_name, header_fmt)

            # largura + alinhamento à direita nas colunas alvo
            for col_idx, col_name in enumerate(df.columns):
                serie_str = df[col_name].astype(str)
                max_len = max(serie_str.map(len).max() if not df.empty else 0, len(str(col_name)))
                largura = max_len + 3   # ≈ +20 px
                if col_name in colunas_direita:
                    worksheet.set_column(col_idx, col_idx, largura, right_fmt)
                else:
                    worksheet.set_column(col_idx, col_idx, largura)

    print(f"✅ Arquivo salvo em: {caminho_arquivo}")

# salva em todos os destinos
for caminho in caminhos_arquivos:
    salvar_excel(caminho)

```
